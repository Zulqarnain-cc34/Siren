# Blurring

```python
f=misc.face()
f=ndimage.gaussian_filter(f,sigma=1)

plt.imshow(f)
plt.show()
```

```python
im = np.zeros((256, 256))
im[64:-64, 64:-64] = 1
im[90:-90,90:-90] = 2
im = ndimage.gaussian_filter(im, 8)

plt.imshow(im)
plt.show()
```

```python
sx=ndimage.sobel(im,axis=0,mode='constant')
sy=ndimage.sobel(im,axis=1,mode='constant')

edges=np.hypot(sx,sy)
plt.imshow(edges)
plt.show()
```

```python
sx=ndimage.laplace(edges)

plt.imshow(sx)
plt.show()
```
